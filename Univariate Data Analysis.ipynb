### Step 4 : Data Visualization
### Univariate - Analysis of single column at a time

```python
A['STATE'].value_counts()
```

```python
sns.countplot(data=A, x='STATE')
plt.title('Countplot for State')
plt.show()
```

```python
for i in cat:
    sns.countplot(data=A, x=i)
    plt.title(f'Countplot for {i}')
    plt.show()
```

### kde - Kernel Density Estimate, it is smoothened version of histogram

```python
sns.histplot(data=A, x='RND', kde=True)
plt.title('Histogram For RND')
plt.show()
```

```python
for i in con:
    sns.histplot(data=A, x=i, kde=True)
    plt.title(f'Histogram for  {i}')
    plt.show()
```

### Bivariate 
1. Con vs Con - Scatterplot
2. Cat vs Con - Boxplot
3. Cat vs cat - crosstab heatmap

```python
con
```

```python
for i in con:
    if i!='PROFIT':
        sns.scatterplot(data=A, x=i, y='PROFIT')
        plt.title(f'Scatterplot for {i} vs PROFIT')
        plt.show()
```
